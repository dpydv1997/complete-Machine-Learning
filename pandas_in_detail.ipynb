### Assignment fetch teh data from url..

### url = 'https://api.github.com/repos/pandas-dev/pandas/issues'

```python
url = 'https://api.github.com/repos/pandas-dev/pandas/issues'
df = pd.read_json(url)
df.head()
```

```python
data = pd.DataFrame(df, columns=['number', 'url','title', 'user', 'labels'])
data
```

```python
id_list = []
node_id = []
html_url = []
user_url_list = []
data_dict = {}
for i in data['user']:
#     print(i.get('url'))
    user_url_list.append(i.get('url'))
    id_list.append(i.get('id'))
    node_id.append(i.get('node_id'))
    html_url.append(i.get('html_url'))
data_dict['user_id'] = id_list
data_dict['node_id'] = node_id
data_dict['user_url'] = user_url_list
data_dict['html_url'] = html_url
print(data_dict)

data1 = pd.DataFrame(data=data_dict)
data1
```

```python
label_node_id = []
label_id = []
label_url = []
color = []
description = []
label_dict = {}
for i in data['labels']:
    
    for j in i:
        label_node_id.append(j.get('node_id'))
        label_id.append(j.get('id'))
        label_url.append(j.get('url'))
        color.append(j.get('color'))
        description.append(j.get('description'))
    
label_dict['label_id'] = label_id
label_dict['label_node_id'] = label_node_id
label_dict['label_url'] = label_url
label_dict['label_id'] = color
label_dict['description'] = description

label_data = pd.DataFrame(label_dict)
label_data
```

```python
df = [data[['number', 'url']], data1, label_data]
df1 = pd.concat(df, axis=1)
```
